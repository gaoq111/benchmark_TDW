```python
from plot_utils import write_images_to_video
import glob
import os
```

```python
positions = ['center', 'top', 'bottom', 'left', 'right']
for position in positions:
	path = f"/data/shared/sim/benchmark/benchmark_TDW/image_capture/{position}"
	names = ["box_room_2018", "empty_scene", "suburb_scene_2018", "archviz_house", "monkey_physics_room", "ruin"]
	output_path = f"/data/shared/sim/benchmark/benchmark_TDW/output_videos/{position}"
	os.makedirs(output_path, exist_ok=True)
	for name in names:
		image_folders = os.path.join(path, name)
		if os.path.exists(image_folders):
			for folder in os.listdir(image_folders):
				folder_path = os.path.join(image_folders, folder)
				tmp = os.listdir(folder_path)
				tmp = sorted([os.path.join(folder_path, i) for i in tmp])
				write_images_to_video(tmp, os.path.join(output_path, f'{name}_{folder}.mp4'))
```
